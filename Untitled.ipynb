```python
import pandas as pd
import numpy as np

#Checking the range of the column to fill in the empty places
df = pd.read_csv('3498906.csv')

column_name = 'FMTM'

min_value = int(df[column_name].min())
max_value = int(df[column_name].max())

print(f"The range of {column_name} is from {min_value} to {max_value}")
```

```python
import pandas as pd

#Checking the range of the column to fill in the empty places
df = pd.read_csv('3498906.csv')

column_name2 = 'PGTM'

min_value2 = int(df[column_name2].min())
max_value2 = int(df[column_name2].max())

print(f"The range of {column_name2} is from {min_value2} to {max_value2}")
```

```python
# Find the indices of empty rows in the specified column
empty_rows = df[df[column_name].isnull()].index

# Fill the empty rows with random values within the range
df.loc[empty_rows, column_name] = np.random.randint(min_value, max_value+1, len(empty_rows))
df[column_name] = df[column_name].astype(int)

# Save the modified DataFrame back to a CSV file
df.to_csv('modified_file.csv', index=False)
```

```python
column_name3 = 'TAVG'

# Define a custom function to fill empty rows with the formula result
def custom_formula(row):
    if pd.isnull(row[column_name3]):
        return (row['TMAX'] + row['TMIN']) / 2
    else:
        return row[column_name3]

# Apply the custom function to fill empty rows with the formula result
df[column_name3] = df.apply(custom_formula, axis=1)

# Save the modified DataFrame back to a CSV file
df.to_csv('modified_file.csv', index=False)
```
